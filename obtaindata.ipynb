```python
import praw
reddit = praw.Reddit(client_id= config["REDDIT"]["client_id"],
                     client_secret= config["REDDIT"]["client_secret"],
                     user_agent=  config["REDDIT"]["user_agent"])

text = []
for submission in reddit.subreddit('all').search('MSFT', time_filter="month", sort="relevance"):
    print(submission.title)
    text.append(submission.title)
#     submission.comments.replace_more(limit=0)
#     comments = submission.comments.list()
#     for comment in comments:
#         print("\t", comment.body)
#         text.append(comment.body)
        
```

```python
import twitter
twit = twitter.Api(consumer_key=config["TWITTER"]["consumer_key"],
            consumer_secret=config["TWITTER"]["consumer_secret"],
            access_token_key=config["TWITTER"]["access_token_key"],
            access_token_secret=config["TWITTER"]["access_token_secret"])

results = twit.GetSearch(raw_query="q=MSFT&result_type=recent&count=100")

for r in results:
    text.append(r.text)
    print(r.text)
```

```python
from aylienapiclient import textapi

aylien = textapi.Client(config["AYLIEN"]["app_id"], config["AYLIEN"]["apikey"])

sentiment = aylien.Sentiment({'text': text})
print(sentiment)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
total = 0
for sentence in text:
#     print(sentence)
    ss = sid.polarity_scores(sentence)
    total += ss["compound"]
    
print(total)
#     for k in ss:
#         print("{0}: {1}, ".format(k, ss[k]), end="")
#         print()
```
